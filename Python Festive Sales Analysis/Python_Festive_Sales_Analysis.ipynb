# Exporatory Data Analysis

```python
#Exporatory Data Analysis (EDA) - examining, visualizing, and summarizing data
```

### Gender

```python
ax=sns.countplot(x='Gender', data=df, hue='Gender', palette='Set2') #bar chart with count frequency

for bars in ax.containers:
    ax.bar_label(bars) #for data labels
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#It groups the data by 'Gender', 
#adds up the 'Amount' for each gender, 
#and then sorts the result from highest to lowest 'Amount'
```

```python
Sex_Wise_Sales= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=Sex_Wise_Sales, hue='Gender')
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men***

### Age Group

```python
ax=sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Age_Grp_Sales = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=Age_Grp_Sales, hue='Age Group')
```

***From above graphs we can see that most of the buyers are of age group between 26-35 are female***

### State

```python
State_Sales = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15,6))
sns.barplot(x='State', y='Orders', data=State_Sales, hue='State')
plt.xticks(rotation=45)
plt.show()
```

```python
# total amount sales from top 10 states
State_Sales_Amt=df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(15,6))
sns.barplot(x='State', y='Amount', data=State_Sales_Amt, hue='State')
plt.xticks(rotation=45)
plt.show()
```

***From above graphs we can see that most of the orders and total sales amount are from UP, Maharashtra and Karnataka respectively***

### Marital Status

```python
ax = sns.countplot(x='Marital_Status', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
MS_Sales=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(5,5))
sns.barplot(x='Marital_Status', y='Amount', data=MS_Sales, hue='Gender')
plt.show()
```

***From above graphs we can see that most of the buyers arre married (women) and they have high purchasing power.***

### Occupation

```python
plt.figure(figsize=(20,5))
ax = sns.countplot(x='Occupation', data=df, hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
O_Sales=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(20,5))
sns.barplot(x='Occupation', y='Amount', data=O_Sales, hue='Occupation')
plt.show()
```

***From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector***

### Product Category

```python
df.columns
```

```python
plt.figure(figsize=(20,6))
ax=sns.countplot(x='Product_Category', data=df, hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45)
plt.show()
```

```python
PC_Sales=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(20,6))
sns.barplot(x='Product_Category', y='Amount', data=PC_Sales, hue='Product_Category')
plt.xticks(rotation=45)
plt.show()
```

***From above graphs we see that most of products are sold from Food, Clothing and Electronics Gadgets***

### Product_ID wise Orders

```python
PID=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(20,5))
sns.barplot(x='Product_ID', y='Orders', data=PID, hue='Product_ID')
plt.show()
```

### Conclusion:

***Married Women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from food, clothing and electronics category***
